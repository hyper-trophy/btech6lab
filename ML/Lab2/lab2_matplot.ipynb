```python
import numpy as np 
np.random.seed(32) 
data = np.random.randint(1,100,50) 
pt.hist(data,rwidth=0.95) 
pt.show() 
```

```python
pt.figure(figsize=(6,6)) # width and height

np.random.seed(16)

data = np.random.randint(1,100,5)

cols = ['red','blue', 'green', 'orange', 'yellow' ]

lab = ["study", 'play', 'gaming', 'sleeping', 'eating' ]

pt.pie(data, colors=cols,labels=lab, autopct="%0.1f%%" ,explode = (0,0.3,0,0,0),textprops={'fontsize':10,'color':'white'}) #maximum value of explode is 1
pt.show()
```
